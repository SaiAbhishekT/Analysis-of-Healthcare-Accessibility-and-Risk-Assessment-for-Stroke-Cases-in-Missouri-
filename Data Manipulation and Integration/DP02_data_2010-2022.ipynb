```python
df = pd.read_csv('ACSDP5Y2022.DP02-Data.csv' , header = 1)
df.head()
```

```python
cols = [col for col in df.columns if 'Margin of Error' not in col]
cols = [col for col in cols if 'Unnamed' not in col]
cols
#df['Margin of Error' in df.columns]
```

```python
comb_df = pd.DataFrame()
flag = 0 
for i in file_list:
    if 'DP02-Data.csv' in i:
        #print(i[7:11], ':' , '','\n')
        df = pd.read_csv(i , header = 1)
        print(i[7:11], ':' , df.shape,'\n')
        #col_name = 'Column Name'+'_'+i[7:11]
        #df.rename({'Column Name' : col_name},axis = 1 , inplace = True)
        df['Year'] =  i[7:11]
        cols = [col for col in df.columns if 'Margin of Error' not in col]
        cols = [col for col in cols if 'Unnamed' not in col]
        df = df[cols]
        
        flag+=1
        if (flag == 1):
            comb_df = df.copy()
        else:
            #comb_df = pd.merge(comb_df , df , on = 'Label' , how ='left')
            comb_df = pd.concat([comb_df , df ] ,axis = 0, sort = False)
```
